```python
feature_series = pd.DataFrame(feature_names, columns=['name'])
feature_series['group'] = feature_series['name'].apply(lambda x: x[0])
```

```python
feature_series_grouped = feature_series.groupby('group').apply(lambda x: x['name'].tolist())
```

```python
def cumdistplot(vec, ax):
    ax.hist(vec, bins=100, density=True, log=False, cumulative=True, fill=False,histtype='step')
    ax.twinx().hist(vec, bins=100, density=True, log=True, cumulative=False, fill=False,histtype='bar')

def grouped_cumdistplot(group_index):
    group = feature_series_grouped.iloc[group_index]
    group_len = len(group)    
    block_size = 4
    n_blocks = int(np.ceil(group_len/block_size))
    fig, axes = plt.subplots(n_blocks,block_size, figsize=(block_size*5,n_blocks*5))
    for index, column_name in enumerate(group):
        row = int(np.floor(index/block_size))
        col = int(index%block_size)
        ax = axes[row][col] if int(np.ceil(group_len/block_size)) >1 else axes[col]
        cumdistplot(features_df[column_name],ax)
        ax.set_title(column_name)
```
